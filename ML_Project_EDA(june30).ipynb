# Bank Marketing
## Problem Statement: 
The data is related with direct marketing campaigns (phone calls) of a Portuguese banking institution. The classification goal is to predict if the client will subscribe a term deposit (variable y).
## Data Set Information: 
The data contains customers features like demographics and transaction history, the bank will be able to predict customer saving behaviours and identify which type of customers is more likely to make term deposits. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or ('no') subscribed.

### Outlier Removal

```python
def remove_outlier(df_in, col_name):
    q1=df_in[col_name].quantile(0.25)
    q3=df_in[col_name].quantile(0.75)
    iqr=q3-q1
    fence_low=q1-1.5*iqr
    fence_high=q3+1.5*iqr
    df_out=df_in.loc[(df_in[col_name]>fence_low)&(df_in[col_name] < fence_high)]
    df_out
    
```

# Feature Engineering

Feature Engineering & Categorical Encoding
Having carried out a brief exploration into the dataset,
let us now proceed onto the task of Feature engineering and numerically encoding the categorical values in our dataset. Feature engineering in a nutshell involves creating new features and relationships from the current features that we have. 


### Numerical columns from categorical columns via the use of the dtype method as follows:

```python
#bank = bank.drop(['response_num'], axis=1)
bank = bank.drop(['duration_min'], axis=1)
categorical=[]
for col,value in bank.iteritems():
    if value.dtype=='object':
        categorical.append(col)
numerical=bank.columns.difference(categorical)     
numerical
```

```python
bank_cat=bank[categorical]
bank_cat=bank_cat.drop(['response'],axis=1)
```

```python
bank_cat=pd.get_dummies(bank_cat)
```

```python
bank_num=bank[numerical]
```

### Combining both numerical and categorical variables

```python
final_data=pd.concat([bank_num,bank_cat],axis=1)
final_data.info()
```

### Target Variable

```python
# Define a dictionary for the target mapping
target_map = {'yes':1, 'no':0}
# Use the pandas apply method to numerically encode our attrition target variable
target = bank["response"].apply(lambda x: target_map[x])
```

## Feature Importance Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(oob_score=True,random_state=0)

model.fit(final_data.drop(columns='response_num'), final_data.response_num)

print(model.feature_importances_)
    
    
col_names = final_data.drop(columns='response_num').columns

importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

std = np.std([tree.feature_importances_ for tree in model.estimators_],
             axis=0)

feat=np.zeros(10)
imp=np.zeros(10)

print("Feature ranking:")

for f in range(0,10):
    print("%d. %s feature %d (%f)" % (f + 1,col_names[f], indices[f], importances[indices[f]]))
    feat[f] = indices[f]
    imp[f] = importances[indices[f]]

# Plot the feature importances
    
plt.title("Feature importances")
plt.bar(feat, imp, color="r", align="center")
```

Here Random Forest is used to select the best features,they are ranked by giving weightage value to each feature and top 10 features are selected whose weightage is high
